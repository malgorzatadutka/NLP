```python
import urllib.request as req
from bs4 import BeautifulSoup
```

```python
#articleURL = "https://www.onet.pl/informacje/onetwiadomosci/polskie-rzeki-przekraczaja-stany-alarmowe-ostrzezenia-najwyzszego-stopnia/xcj7w8y,79cfc278"
articleURL = "https://www.washingtonpost.com/climate-solutions/2024/01/20/heating-tips-winter-space-heater-fireplace/"
```

```python
def getArticle(url):
    page = req.urlopen(url).read().decode('utf8', 'ignore') # download the webpage
    soup = BeautifulSoup(page, "lxml") 
    text = ''.join(map(lambda p: p.text, soup.find_all('article')))
    return text

texting = getArticle(articleURL)
texting
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sentences = sent_tokenize(text) # split text into sentences
sentences
```

```python
word_in_senteces = word_tokenize(text.lower()) # split senteces into words
word_in_senteces
```

```python
custom_stop_words = set(stopwords.words('english') + list(punctuation)) 
words_wo_stop_words = [word.lower() for word in word_tokenize(text) if word.lower() not in custom_stop_words] # remove stopwords
words_wo_stop_words
```

```python
from nltk.probability import FreqDist
frequency = FreqDist(words_wo_stop_words) # count how many times word is present in article
frequency
```

```python
from heapq import nlargest
nlargest(15, frequency, key=frequency.get) # find the most common words
```

```python
from collections import defaultdict # it doesn't throw an error if the key is not present, insead it adds that key
ranking = defaultdict(int) # count the importance of the particular sentence based on the word frequency
for i, sentence in enumerate(sentences):
    for word in word_tokenize(sentence.lower()):
        if word in frequency:
            ranking[i] += frequency[word]
ranking
```

```python
sentences_indexes = sorted(nlargest(3, ranking, key=ranking.get)) # show indexes of the most important sentences
sentences_indexes
```

```python
[sentences[j] for j in sentences_indexes]
```

```python
def summarizeText(text, n):
    sentences = sent_tokenize(text)
    assert n <= len(sentences)
    word_in_senteces = word_tokenize(text.lower())
    custom_stop_words = set(stopwords.words('english') + list(punctuation)) 
    words_wo_stop_words = [word.lower() for word in word_tokenize(text) if word.lower() not in custom_stop_words]
    frequency = FreqDist(words_wo_stop_words)
    ranking = defaultdict(int) # count the importance of the particular sentence based on the word frequency
    for i, sentence in enumerate(sentences):
        for word in word_tokenize(sentence.lower()):
            if word in frequency:
                ranking[i] += frequency[word]
    sentences_indexes = sorted(nlargest(n, ranking, key=ranking.get))
    return [sentences[j] for j in sentences_indexes]
```

```python
summarizeText(text, 3)
```
